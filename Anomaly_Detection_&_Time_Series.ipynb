```python
'''
1.What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.
        -Anomaly Detection (also called outlier detection) is a data  analysis technique used to identify patterns, events, or observations that deviate significantly from the normal behavior of a dataset.
        -These unusual data points are called anomalies, outliers, or exceptions.
        -Types of Anomaly Detection:
          1.Point Anomalies:
            -A point anomaly occurs when a single data instance is very different from the rest of the data.
            -It is the simplest and most common type of anomaly.
            -Characteristics:
              -Only one data point is abnormal
              -It does not depend on context
              -It stands out clearly from other observations
            -Examples:
              -A credit card transaction of ₹5,00,000 when usual spending is ₹5,000
              -A student scoring 5 marks out of 100 when average is 85
          2.Contextual Anomalies:
            -A contextual anomaly is a data point that is abnormal only in a specific context.
            -Here, the same value may be normal in one situation but abnormal in another.
            -Characteristics:
              -Depends on time, location, or other contextual information.
              -Requires additional attributes to determine abnormality.
            -Example:
              -High electricity usage at midnight (abnormal)
              -Large network traffic at 3 AM
          3.Collective Anomalies
            -A collective anomaly occurs when a group of related data points together behave abnormally,
            -even though individual points may appear normal.
            -Characteristics:
              -An anomaly is detected in a sequence or group
              -Individual values may not be anomalous
              -The pattern as a whole is unusual
            -Examples:
              -A sequence of failed login attempts on a server
              -Sudden drop in sales for several consecutive days
'''
```

```python
'''
2. Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.
    -All three are popular methods for detecting anomalies.
    -But they differ greatly in their approach, assumptions, and suitable applications.
    1. Isolation Forest:
    -Approach:
        -Isolation Forest is a tree-based ensemble algorithm.
        -It isolates anomalies instead of profiling normal data.
        -Works on the principle: Anomalies are few and different → therefore easier to isolate.
    -How It Works:
        -Randomly selects a feature and split value.
        -Builds many random trees.
        -Points that require fewer splits to isolate are considered anomalies.
    -Key Characteristics:
        -Works well with high-dimensional data
        -Does not rely on distance or density
        -Fast and scalable
        -Robust to irrelevant features
    -Use Cases:
        -Fraud detection
        -Network intrusion detection
        -Large-scale datasets
    2.DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
    -Approach:
        -DBSCAN is a density-based clustering algorithm.
        -It groups dense regions and labels sparse points as anomalies.
    -How It Works:
        -Uses two parameters:ε (epsilon): neighborhood radius and minPts: minimum points required to form a cluster
        -Points are classified as:Core points, Border points and Noise points (anomalies)
    -Key Characteristics:
        -Does not require number of clusters
        -Can detect clusters of arbitrary shape
        -Naturally identifies outliers as “noise”
    -Use Cases:
        -Geospatial data
        -Spatial pattern detection
        -Image processing
    3.Local Outlier Factor (LOF)
    -Approach:
        -LOF is a density-based outlier detection algorithm.
        -Detects anomalies by comparing local density of a point with its neighbors.
    -How It Works:
        -Computes a score called LOF score:
          -LOF ≈ 1 → normal
          -LOF > 1 → possible anomaly
          -Higher LOF → stronger anomaly
        -Key Characteristics:
          -Captures local anomalies effectively
          -Good for datasets with:
        -Use Cases:
          -Credit card fraud
          -Credit card fraud
          -Network traffic analysis
          -Intrusion detection
'''
```

```python
'''
3.What are the key components of a Time Series? Explain each with one example.
  -A Time Series is a sequence of data points collected or recorded at regular time intervals (daily, monthly, yearly, etc.).
  -Examples: stock prices, temperature readings, sales data.
  -Every time series is generally composed of four key components:
      -Trend
      -Seasonality
      -Cyclic Component
      -Irregular (Random/Noise) Component
  1.Trend Component:
  -The trend represents the long-term movement or general direction of the data over time.
  -It shows whether the data is:Increasing, Decreasing or Remaining stable
  -Trend ignores short-term fluctuations and focuses on overall behavior.
  -Characteristics:
    -Slow and gradual movement
    -Exists over a long period
    -Can be upward, downward, or constant
  -Example: Annual sales of smartphones: If smartphone sales increase steadily from 2015 to 2024
            due to growing demand and technology adoption, this steady rise represents a trend component.
  2. Seasonal Component:
  -Seasonality refers to patterns that repeat at regular and fixed time intervals such: Daily, weekly, monthly, annually.
  -These variations occur due to seasonal factors.
  -Characteristics:
    -Short-term and periodic
    -Repeats at fixed intervals
    -Predictable in nature
  -Example:Ice cream sales: Higher during summer and lower in winters.
 3. Cyclic Component:
 -The cyclic component refers to fluctuations in data that occur over long, irregular periods.
 -Unlike seasonality, cycles:Do not have a fixed time period and Depend on economic or business conditions.
 -Characteristics:
    -Long-term oscillations
    -Not strictly periodic
    -Often influenced by external factors
-Example: Business or Economic Cycles: Boom period → high sales and Recession → low sales.
          These ups and downs in the economy occurring over several years represent cyclic behavior.
4. Irregular (Random/Noise) Component
-The irregular component consists of unpredictable and random variations in the time series.
-These are sudden changes that cannot be explained by trend, seasonality, or cycles.
-Characteristics:
    -Random and unexpected
    -Short-term
    -Cannot be forecasted easily
-Example: Sudden Drop in Sales Due to Natural Disaster
          If a flood or pandemic suddenly reduces sales for a few months, that effect is an irregular component.
'''
```

```python
'''
4. Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?
    -A time series is said to be stationary if its statistical properties remain constant over time.
    -A stationary time series has no trend, no seasonality, and constant variance over time.
    -How to Test Stationarity?
      -There are two main approaches:
      1. Visual Inspection (Graphical Method):
      Plot the time series and check for:
       - Upward or downward trend
       - Changing variance
        -Seasonal patterns
        -If any of these exist → series is likely non-stationary.
      2. Statistical Tests:
      (A) Augmented Dickey–Fuller (ADF) Test: Most commonly used test.
      -Null Hypothesis (H₀):Series is non-stationary (has unit root)
      -Alternative Hypothesis (H₁):Series is non-stationary (has unit root)
      -Alternative Hypothesis (H₁):Series is stationary
      -Decision Rule: p-value < 0.05 → Reject H₀ → Stationary, p-value ≥ 0.05 → Fail to reject H₀ → Non-stationary
      (B) KPSS Test: Opposite of ADF
      -Null Hypothesis: Series is stationary
      -Alternative: Series is non-stationary
      -Used as a complementary test to ADF.
    -How to Transform a Non-Stationary Series into Stationary?
      -If a series is non-stationary, we apply transformation techniques.
      1. Differencing: Most common method.
      -First Order Differencing:Y′t​=Yt​−Yt−1​
      -Removes trend from the series.
      -If still non-stationary → apply second differencing.
      2. Log Transformation: Used when variance increases with time.
      -Helps stabilize variance.:Y′t​=log(Yt​)
      3. Detrending:Remove deterministic trend using regression.
      -Yt′​=Yt​−Y^t​
      4.Seasonal Differencing: Used when seasonality is present.
      -Yt′​=Yt​−Yt−s​, where s = seasonal period (e.g., 12 for monthly data)
      5. Transformation + Differencing:Often a combination works best:
      -Log + Differencing
      -Square root + Differencing
'''
```

```python
'''
5.  Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.
      -All these models belong to classical time series forecasting techniques.
      -They are used to model and predict future values based on past data.
      -Each successive model is an extension of the previous one.
      1. AR Model – Autoregressive Model
        -Structure:Yt​=c+ϕ1​Yt−1​+ϕ2​Yt−2​+...+ϕp​Yt−p​+ϵt​
        -Current value depends on its own previous values.
        -Parameters:p → number of past observations used
        -Key Idea:Future value is a regression on its own past.
        -Application: Used when:Data is stationary and Strong correlation exists with past values.
        -Examples: Forecasting temperature from previous days, Strong correlation exists with past values.
      2. MA Model – Moving Average Model
        -Structure: Yt=c+ϵt+θ1ϵt−1+...+θqϵt−q
        -Current value depends on past errors.
        -Parameters:q → numberof lagged error terms
        -Key Idea:Captures short-term random fluctuations.
        -Application: Used when:Series is stationary and Random shocks strongly affect the data.
        -Examples:Demand forecasting with random variations, Modeling noise in financial data
      3. ARIMA Model-AutoRegressive Integrated Moving Average
        -Structure: ARIMA(p,d,q), Where: p=AR order, d=differencing order, q=MA order
        -Key Idea:Used for non-stationary data.
                -Differencing (d) converts non-stationary series into stationary
                -Combines AR and MA components
	​      -Application:Used when:Data has trend and No seasonality
        -Examples: Stock price forecasting, GDP or revenue prediction with trend
      4. SARIMA Model-Seasonal AutoRegressive Integrated Moving Average
        -Structure: SARIMA(p,d,q)(P,D,Q)s, Where:p,d,q → non-seasonal ARIMA terms, P,D,Q → seasonal ARIMA terms, s → seasonal period
        -Key Idea:Handles:Trend, Seasonality and Autocorrelation
        -Application: Used when:Data shows repeating seasonal patterns
        -Examples: Airline passenger traffic, Monthly electricity demand, Festival sales forecasting
      5. SARIMAX Model-Seasonal AutoRegressive Integrated Moving Average with Exogenous Variables
        -Structure: SARIMAX(p,d,q)(P,D,Q)s+X, Extends SARIMA by including external (independent) variables
        -Key Idea: Future values depend on:Past values, Seasonal patterns and External influencing factors
        -Used when:Target variable is affected by external factors
        -Examples:Sales depending on: advertising spend, holidays and discounts
  '''
```

```python
#7. Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.
# Step 1: Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Step 2: Create a sample dataset similar to NYC Taxi Fare
# (trip_distance vs fare_amount)

np.random.seed(42)

# Generate normal data
trip_distance = np.random.normal(10, 2, 200)      # average 10 miles
fare_amount = trip_distance * 2.5 + np.random.normal(0, 3, 200)

# Step 3: Add some artificial anomalies
trip_distance_outliers = np.array([2, 50, 1, 60, 55])
fare_outliers = np.array([150, 5, 200, 10, 180])

trip_distance = np.concatenate([trip_distance, trip_distance_outliers])
fare_amount = np.concatenate([fare_amount, fare_outliers])

# Step 4: Create DataFrame
df = pd.DataFrame({'trip_distance': trip_distance,'fare_amount': fare_amount})

print("First 5 rows of dataset:")
print(df.head())

# Step 5: Apply Isolation Forest
model = IsolationForest(contamination=0.05, random_state=42)

# Fit the model and predict anomalies
df['anomaly'] = model.fit_predict(df[['trip_distance', 'fare_amount']])

# Step 6: Separate normal points and anomalies
normal_points = df[df['anomaly'] == 1]
anomalous_points = df[df['anomaly'] == -1]

print("\nTotal data points:", len(df))
print("Normal points:", len(normal_points))
print("Anomalies detected:", len(anomalous_points))

# Step 7: Visualize anomalies using 2D scatter plot
plt.figure(figsize=(8,6))

plt.scatter(normal_points['trip_distance'],
            normal_points['fare_amount'],
            label='Normal Points',
            color='blue')

plt.scatter(anomalous_points['trip_distance'],
            anomalous_points['fare_amount'],
            label='Anomalies',
            color='red')

plt.title("Anomaly Detection using Isolation Forest")
plt.xlabel("Trip Distance")
plt.ylabel("Fare Amount")
plt.legend()
plt.show()

# Step 8: Display detected anomalies
print("\nDetected Anomalies:")
print(anomalous_points)
```

```python
#8. Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.
# Step 1: Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Step 2: Load the AirPassengers dataset
data = pd.read_csv("AirPassengers.csv")

# Step 3: Convert Month column to datetime and set as index
data['Month'] = pd.to_datetime(data['Month'])
data.set_index('Month', inplace=True)

# Step 4: Plot the original time series
plt.figure(figsize=(10,5))
plt.plot(data['Passengers']) # Corrected column name
plt.title("Original AirPassengers Data")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.show()

# Step 5: Train SARIMA model
# Using commonly accepted parameters for this dataset
model = SARIMAX(
    data['Passengers'], # Corrected column name
    order=(1, 1, 1),               # ARIMA part
    seasonal_order=(1, 1, 1, 12),  # Seasonal part
    enforce_stationarity=False,
    enforce_invertibility=False
)

sarima_model = model.fit(disp=False)

# Step 6: Print model summary (optional)
print(sarima_model.summary())

# Step 7: Forecast next 12 months
forecast = sarima_model.forecast(steps=12)

# Step 8: Create date index for forecasted values
future_dates = pd.date_range(start=data.index[-1], periods=13, freq='M')[1:]

forecast_series = pd.Series(forecast.values, index=future_dates)

# Step 9: Visualize actual data and forecast
plt.figure(figsize=(10,5))
plt.plot(data['Passengers'], label='Actual Data') # Corrected column name
plt.plot(forecast_series, label='Forecast', color='red')
plt.title("SARIMA Forecast for Next 12 Months")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()

# Step 10: Display forecasted values
print("Forecast for Next 12 Months:")
print(forecast_series)
```

```python
#9.Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.
# Step 1: Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# Step 2: Create a sample numerical dataset
# (Example similar to real-world numeric data)

np.random.seed(42)

# Generate normal data (two related numerical features)
x = np.random.normal(50, 5, 200)
y = x + np.random.normal(0, 3, 200)

# Add some artificial anomalies
x_outliers = np.array([20, 80, 10, 90, 5])
y_outliers = np.array([100, 10, 120, 5, 110])

x = np.concatenate([x, x_outliers])
y = np.concatenate([y, y_outliers])

# Step 3: Create DataFrame
df = pd.DataFrame({
    'Feature1': x,
    'Feature2': y
})

print("Dataset Sample:")
print(df.head())

# Step 4: Apply Local Outlier Factor (LOF)

lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)

df['anomaly'] = lof.fit_predict(df[['Feature1', 'Feature2']])

# LOF returns:
#  1  -> Normal point
# -1  -> Anomaly

# Step 5: Separate normal points and anomalies
normal = df[df['anomaly'] == 1]
anomalies = df[df['anomaly'] == -1]

print("\nTotal Points:", len(df))
print("Normal Points:", len(normal))
print("Anomalies Detected:", len(anomalies))

# Step 6: Visualize the results

plt.figure(figsize=(8,6))

plt.scatter(normal['Feature1'], normal['Feature2'],
            label='Normal Points', color='blue')

plt.scatter(anomalies['Feature1'], anomalies['Feature2'],
            label='Anomalies', color='red')

plt.title("Anomaly Detection using Local Outlier Factor (LOF)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()

# Step 7: Display detected anomalies
print("\nDetected Anomalies:")
print(anomalies)
```

```python
# ===============================================
# POWER GRID MONITORING – COMPLETE WORKFLOW CODE
# ===============================================

# 1. Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.ensemble import IsolationForest
from sklearn.metrics import mean_absolute_error, mean_squared_error

from statsmodels.tsa.statespace.sarimax import SARIMAX


# ------------------------------------------------
# 2. CREATE / LOAD DATASET
# (Replace this with real dataset in practice)
# ------------------------------------------------

np.random.seed(42)

# Simulating 15-minute interval data
date_range = pd.date_range(start="2024-01-01", periods=1000, freq="15T")

energy_usage = 200 + 20*np.sin(np.arange(1000)/50) + np.random.normal(0, 5, 1000)

temperature = 25 + 5*np.sin(np.arange(1000)/100) + np.random.normal(0, 2, 1000)

humidity = 60 + 10*np.cos(np.arange(1000)/80) + np.random.normal(0, 3, 1000)

df = pd.DataFrame({
    "timestamp": date_range,
    "energy_usage": energy_usage,
    "temperature": temperature,
    "humidity": humidity
})

print("Sample Data:")
print(df.head())


# ------------------------------------------------
# 3. PREPROCESSING
# ------------------------------------------------

df["timestamp"] = pd.to_datetime(df["timestamp"])
df.set_index("timestamp", inplace=True)

# Feature Engineering
df["hour"] = df.index.hour
df["dayofweek"] = df.index.dayofweek

# Handle missing values
df.fillna(method="ffill", inplace=True)


# ------------------------------------------------
# 4. ANOMALY DETECTION USING ISOLATION FOREST
# ------------------------------------------------

print("\nRunning Anomaly Detection...")

features = df[["energy_usage", "temperature", "humidity"]]

iso_model = IsolationForest(contamination=0.02, random_state=42)

df["anomaly"] = iso_model.fit_predict(features)

# Extract anomalies
anomalies = df[df["anomaly"] == -1]

print("Total Anomalies Detected:", len(anomalies))


# Visualization of anomalies
plt.figure(figsize=(12, 5))

plt.plot(df.index, df["energy_usage"], label="Energy Usage")

plt.scatter(anomalies.index, anomalies["energy_usage"],
            color="red", label="Anomalies")

plt.title("Real-Time Energy Usage Anomaly Detection")
plt.xlabel("Time")
plt.ylabel("Energy Usage")
plt.legend()
plt.show()


# ------------------------------------------------
# 5. TIME SERIES FORECASTING USING SARIMAX
# ------------------------------------------------

print("\nTraining SARIMAX Model for Forecasting...")

y = df["energy_usage"]
X = df[["temperature", "humidity"]]

# Train-Test Split
train_size = int(len(df) * 0.8)

y_train, y_test = y[:train_size], y[train_size:]
X_train, X_test = X[:train_size], X[train_size:]

# Build SARIMAX Model
sarimax_model = SARIMAX(
    y_train,
    exog=X_train,
    order=(1,1,1),
    seasonal_order=(1,1,1,96),   # 96 intervals = daily seasonality
    enforce_stationarity=False,
    enforce_invertibility=False
)

sarimax_result = sarimax_model.fit(disp=False)

print("\nSARIMAX Model Trained Successfully")


# ------------------------------------------------
# 6. FORECAST ON TEST DATA
# ------------------------------------------------

print("\nGenerating Forecast...")

pred = sarimax_result.predict(
    start=len(y_train),
    end=len(y)-1,
    exog=X_test
)


# ------------------------------------------------
# 7. MODEL VALIDATION
# ------------------------------------------------

mae = mean_absolute_error(y_test, pred)
rmse = np.sqrt(mean_squared_error(y_test, pred))

print("\nModel Performance:")
print("MAE :", mae)
print("RMSE:", rmse)


# Visualization of forecast vs actual
plt.figure(figsize=(12,5))

plt.plot(y_test.index, y_test, label="Actual Usage")
plt.plot(y_test.index, pred, label="Forecasted Usage", color="green")

plt.title("SARIMAX Forecast vs Actual Energy Demand")
plt.xlabel("Time")
plt.ylabel("Energy Usage")
plt.legend()
plt.show()


# ------------------------------------------------
# 8. REAL-TIME FUTURE FORECAST (NEXT 2 HOURS)
# ------------------------------------------------

print("\nForecasting Next 2 Hours (8 intervals)...")

future_exog = X[-8:]

future_forecast = sarimax_result.forecast(
    steps=8,
    exog=future_exog
)

print("\nNext 2 Hours Forecasted Values:")
print(future_forecast)


plt.figure(figsize=(10,4))

plt.plot(y[-50:], label="Recent Actual Data")
plt.plot(future_forecast, label="Next 2 Hours Forecast", color="orange")

plt.title("Short Term Energy Demand Forecast")
plt.xlabel("Time")
plt.ylabel("Energy Usage")
plt.legend()
plt.show()


# ------------------------------------------------
# 9. MONITORING FORECAST PERFORMANCE OVER TIME
# ------------------------------------------------

error = abs(y_test - pred)

plt.figure(figsize=(10,4))
plt.plot(error, label="Forecast Error")
plt.title("Forecast Error Over Time")
plt.xlabel("Time")
plt.ylabel("Absolute Error")
plt.legend()
plt.show()


# ------------------------------------------------
# 10. RETRAINING FUNCTION (FOR FUTURE USE)
# ------------------------------------------------

def retrain_model(new_data):

    y_new = new_data["energy_usage"]
    X_new = new_data[["temperature", "humidity"]]

    model = SARIMAX(
        y_new,
        exog=X_new,
        order=(1,1,1),
        seasonal_order=(1,1,1,96)
    )

    return model.fit(disp=False)


print("\nWorkflow Execution Completed Successfully!")
```
